```python
from PIL import Image
import numpy as np
# Open an image file
image_path = 'Y184.JPG'
original_image = Image.open(image_path)

# Resize the image
resized_image = original_image.resize((240, 240))

# Convert the image to a NumPy array
resized_array = np.array(resized_image)

print(resized_array.shape)
img = Image.fromarray(resized_array)
display(img)
```

```python
X=[]
y=[]
import os
from os.path import isfile, join
main_dir="brain_tumor_dataset"
for dir in os.listdir(main_dir):
  for image_file in os.listdir(dir):
    print(image_file)
    image_path = main_dir+'/'+dir+'/'+image_file
    original_image = Image.open(image_path)
    original_image = original_image.convert('RGB')
    resized_image = original_image.resize((240, 240))
    resized_array = np.array(resized_image)
    print(resized_array.shape)
    X.append(resized_array)
    if dir=="yes":
      y.append(1)
    else:
      y.append(0)
```

```python
X=np.array(X)
y=np.array(y)
print(X.shape)
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
input_shape = (240, 240, 3)
model = models.Sequential()
model.add(layers.Conv2D(32, kernel_size=(1,1), input_shape=input_shape, activation='relu', strides=(1, 1)))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))
model.add(layers.Flatten(input_shape=(15, 15, 32)))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.summary()
```

```python
X_val=[]
y_val=[]
import os
from os.path import isfile, join
main_dir=["no","yes"]
for dir in main_dir:
  for image_file in os.listdir(dir):
    print(image_file)
    image_path = dir+'/'+image_file
    original_image = Image.open(image_path)
    original_image = original_image.convert('RGB')
    resized_image = original_image.resize((240, 240))
    resized_array = np.array(resized_image)
    print(resized_array.shape)
    X_val.append(resized_array)
    if dir=="yes":
      y_val.append(1)
    else:
      y_val.append(0)
```

```python
def hms_string(seconds):
    hours, remainder = divmod(seconds, 3600)
    minutes, seconds = divmod(remainder, 60)
    return "{:02}:{:02}:{:02}".format(int(hours), int(minutes), int(seconds))
```

```python
import time
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
start_time = time.time()
X_val=np.array(X_val)
y_val=np.array(y_val)
model.fit(x=X, y=y, batch_size=32, epochs=5, validation_data=(X_val, y_val))

end_time = time.time()
execution_time = (end_time - start_time)
print(f"Elapsed time: {hms_string(execution_time)}")
```
